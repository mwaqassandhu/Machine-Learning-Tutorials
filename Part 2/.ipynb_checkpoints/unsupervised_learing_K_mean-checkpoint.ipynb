## K-Nearest Neighbors (KNN) Algorithm Tutorial (by Muhammad Waqas)

- **Overview**: The K-Nearest Neighbors (KNN) algorithm is a simple yet effective machine learning technique for classification and regression tasks.
  
- **Instance-Based Learning**: KNN is an instance-based or lazy learning method, meaning it doesn't build a model during training. Instead, it stores the training data and makes predictions based on the closest data points when needed.

- **How KNN Works**:
  1. **Choose K Value**: Determine the number of neighbors (K) to consider for predictions. Typically choose an odd K to avoid ties.
  
  2. **Calculate Distances**: Compute the distance (e.g., Euclidean, Manhattan, cosine) between the new data point and all training data points.
  
  3. **Find K Nearest Neighbors**: Identify the K data points with the smallest distances, known as the K-nearest neighbors.
  
  4. **Make Predictions**:
     - For classification: Use "majority voting" among the K-nearest neighbors to assign the predicted class.
     - For regression: Average the target values of the K-nearest neighbors to predict the value.
  
- **Key Considerations**:
  - **Distance Metric**: Choose an appropriate distance metric based on the data and problem.
  - **Feature Scaling**: Normalize or scale features to ensure KNN isn't sensitive to feature scales.
  - **Choosing K**: Carefully select K to balance model flexibility and bias. The value of K should be carefully selected, as it affects the trade-off between bias and variance. Smaller values of K (e.g., 1) result in a more flexible, lower-bias model, but it can be noisy. Larger values of K (e.g., 10 or 20) lead to a smoother decision boundary but may introduce higher bias.

Now that we've covered the theory of the KNN algorithm, let's proceed with the practical implementation using the `make_blobs` dataset as previously described.

### Step 1: Calculate Distance

```python
from math import sqrt
def calculate_distance(row1, row2):
    return sqrt( sum( [(row1[0][i] - row2[0][i])**2 for i in range(len(row1))]))

# Test distance function
# ([x,y], l) means vector is [x,y] and l is the label. 
dataset = [
 ([1,2], 1), ([2,1], 1),
 ([0,1], 0), ([1,0], 0)
]
# compute distance from a data point
test_row = ([0, 0], -1)
distances = [calculate_distance(test_row, row) for row in dataset]
print(distances)
```

## Get Nearest Neighbors

```python
def get_neighbors(train, test_row, num_neighbors=3):

    #compute distances
    distances = [(train_row, calculate_distance(test_row, train_row))
    for train_row in train]

    #sort
    distances.sort(key=lambda tup: tup[1])

    #get top-k neighbors
    neighbors = [distances[i][0] for i in range(num_neighbors)]

    return neighbors

# test get_get_neighbors function with top-3 neighbors
neighbors = get_neighbors(dataset, test_row)
print(neighbors)
```

## Step 3: Get Prediction

```python
def predict_classification(train, test_row, num_neighbors=3):

    #get top-k neighbors
    neighbors = get_neighbors(train, test_row, num_neighbors)

    # get most frequent label
    output_values = [row[-1] for row in neighbors]
    prediction = max(set(output_values), key=output_values.count)

    return prediction
```

```python
# test prediction function
test_row = ([2, 2], 1)
prediction = predict_classification(dataset, test_row)
print(prediction)
```

# KNN Demo

We will use the `make_blobs` dataset from scikit-learn to demonstrate K-Nearest Neighbors (KNN) clustering. The `make_blobs` function generates random data with clusters, making it suitable for this purpose.

**Step 1: Import Required Libraries**

Let's start by importing the necessary libraries:


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

**Step 2: Generate and Visualize the `make_blobs` Dataset**

We'll create a synthetic dataset using the `make_blobs` function and visualize it.


```python
# Generate a synthetic dataset with three clusters
X, y = make_blobs(n_samples=500, centers=5, random_state=42,center_box=(-10.0, 10.0),)

# Plot the generated dataset
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Generated Blobs Dataset')
plt.show()
```


**Step 3: Split the Dataset and Train-Test Split**

Next, we'll split the dataset into features (X) and the target variable (y). Then, we'll divide it into training and testing sets.


```python
# Split the dataset into features and target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

**Step 4: Build and Train the KNN Model**

Create a KNN classifier and train it using the training data.


```python
# Create a KNN classifier with a specified number of neighbors (e.g., 3)
k = 5
knn_classifier = KNeighborsClassifier(n_neighbors=k)

# Train the classifier on the training data
knn_classifier.fit(X_train, y_train)
```

**Step 5: Make Predictions and Evaluate the Model**

Make predictions on the test data and evaluate the model using metrics like accuracy, confusion matrix, and classification report.


```python
# Make predictions on the test data
y_pred = knn_classifier.predict(X_test)

# Evaluate the model
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nAccuracy Score:", accuracy_score(y_test, y_pred))
```

**Step 6: Choosing the Right Value of K**

Similar to the previous tutorial, you can experiment with different values of K to find the optimal one using techniques like cross-validation.

**Step 7: Visualize the Decision Boundary (Optional)**

You can visualize the decision boundary of the KNN classifier to see how it separates the clusters in the data. This can help you understand the algorithm's behavior.



```python
def plot_decision_boundary(X, y, model, title):
    h = .02  # Step size in the mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
    
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(title)
    plt.show()

# Assuming 'knn_classifier' has already been trained as in the previous steps
plot_decision_boundary(X_train, y_train, knn_classifier, 'KNN Decision Boundary')
```

This tutorial demonstrates K-Nearest Neighbors clustering using the `make_blobs` dataset from scikit-learn. The principles discussed here can be applied to other datasets and problems, helping you gain a better understanding of KNN clustering in practice.
