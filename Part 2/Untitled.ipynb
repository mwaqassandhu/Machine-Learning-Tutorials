In this tutorial, we'll walk through the process of using Support Vector Machines (SVM) to classify a two-dimensional dataset. We'll visualize the decision boundary to gain a better understanding of how SVM works.

### Tutorial: SVM with 2-Dimensional Data Visualization

#### Step 1: Generate Synthetic Data

We'll start by generating a synthetic dataset with two classes for easy visualization.


```python
import numpy as np

# Set random seed for reproducibility
np.random.seed(0)

# Generate random data for class 1
X1 = np.random.randn(20, 2) + [2, 2]

# Generate random data for class 2
X2 = np.random.randn(20, 2) + [-2, -2]

# Generate random data for class 3
X3 = np.random.randn(20, 2) + [-2, 2]

# Combine the data
X = np.vstack([X1, X2, X3])
y = np.hstack([np.full(20, 0), np.full(20, 1), np.full(20, 2)])
```

#### Step 2: Visualize the Data

```python
import matplotlib.pyplot as plt

# Scatter plot for visualizing data
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='blue', label='Class 0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='red', label='Class 1')
plt.scatter(X[y == 2][:, 0], X[y == 2][:, 1], color='green', label='Class 2')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

#### Step 3: Train the SVM

```python
from sklearn.svm import SVC

# Initialize and train the SVM
svm = SVC(kernel='linear', C=1.0)
svm.fit(X, y)
```

#### Step 4: Visualize the Decision Boundary


```python
def plot_decision_boundary(X, y, model, ax):
    # Plot data points
    ax.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='blue', label='Class 0')
    ax.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='red', label='Class 1')
    ax.scatter(X[y == 2][:, 0], X[y == 2][:, 1], color='green', label='Class 2')
    
    # Create mesh grid for plotting decision boundary
    xx, yy = np.meshgrid(np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 500),
                         np.linspace(X[:, 1].min() - 1, X[:, 1].max() + 1, 500))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # Plot decision boundary
    ax.contourf(xx, yy, Z, alpha=0.2, levels=[0, 1, 2], colors=['blue', 'red', 'green'])
    
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.legend()

# Visualize decision boundary
fig, ax = plt.subplots()
plot_decision_boundary(X, y, svm, ax)
plt.show()
```

#### Step 5: Interpretation

You should now see a plot with three classes of data points and decision boundaries created by the SVM. Each region corresponds to a predicted class.

### Conclusion

In this tutorial, we've generated a synthetic 2-dimensional dataset, trained an SVM classifier, and visualized the decision boundary. This provides a clear understanding of how SVM works in a simple scenario. Keep in mind that in real-world applications, the data may not always be linearly separable, and more complex kernels or techniques may be necessary.
