## Load Data
This code loads the Iris dataset, creates a Pandas DataFrame from it, and displays the first few rows of the dataset. Here's what you'll see:

```python
from sklearn.datasets import load_iris

# Load the Iris dataset from scikit-learn
data = load_iris()

# Create a DataFrame from the dataset
df = pd.DataFrame(data.data, columns=data.feature_names)

# Add the target column
df['target'] = data.target

# Display the first few rows of the DataFrame
print(df.head())
```

# Get the shape of the dataset


```python
print("Shape of the dataset:", df.shape)
```

# Basic statistics of numeric columns

```python
print("Statistics of numeric columns:\n", df.describe())
```

# Information about the dataset


```python
print("Information about the dataset:\n", df.info())
```

# Count unique values in the 'target' column


```python
print("Unique values in 'target' column:", df['target'].value_counts())
```

## Data Preprocessing

```python
# Rename columns for readability
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# Filter rows based on a condition (e.g., sepal length > 5.0)
filtered_df = df[df['sepal_length'] > 5.0]
```

## Data Visualization


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pairplot to visualize relationships between numeric columns
sns.pairplot(df, hue='species', palette='Set1')
plt.show()
```
