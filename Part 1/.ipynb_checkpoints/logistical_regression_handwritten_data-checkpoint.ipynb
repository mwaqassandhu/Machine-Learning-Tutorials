```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import tensorflow as tf




# Load the MNIST dataset from TensorFlow/Keras
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
## You can do PCA to reduce the dimensions
# X_train_pca, y_train_pca, X_test_pca, y_test_pca
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# Plot a few images with their labels
plt.figure(figsize=(10, 4))

for i in range(10):  # Plot the first 10 images
    plt.subplot(2, 5, i + 1)
    plt.imshow(X_train[i], cmap='gray')
    plt.title(f"Label: {y_train[i]}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python

# Reshape the data to a 2D format (samples, features)
X_train = X_train.reshape(X_train.shape[0], -1)
X_test = X_test.reshape(X_test.shape[0], -1)
```

```python

# Create and train a logistic regression model
model = LogisticRegression(max_iter=100)
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)
# Calculate the confusion matrix
confusion = confusion_matrix(y_test, y_pred)

# Print the confusion matrix
print("Confusion Matrix:")
print(confusion)

# You can also visualize the confusion matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(confusion, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import classification_report, accuracy_score
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)


# Generate a classification report
report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)
```
